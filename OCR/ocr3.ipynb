```python
import cv2
import pytesseract
img=cv2.imread("ocr3.jpeg")
img
```

```python
cv2.imshow("frame",img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
#set path into programmimg files
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
#extract test from image
data=pytesseract.image_to_string(img)
print(data)
```

```python
dic=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
n=len(dic['text'])
n
```

```python
for i in range(n):
    if dic['conf'][i]>=50:
        x=dic["left"][i]
        y=dic["top"][i]
        w=dic["width"][i]
        h=dic['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)
cv2.imshow("frame",img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
