```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mlt
import seaborn as sns
import tensorflow as tf
df=pd.read_csv('Automobile.csv')
df
```

```python
df['Occupation'].unique()
```

```python
df.drop(['State','Gender','Customer ID','Age','Gender','Hobbies'],axis=1,inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Region']=lab.fit_transform(df['Region'])
df['Occupation']=lab.fit_transform(df['Occupation'])
df['Vehicle_Segment']=lab.fit_transform(df['Vehicle_Segment'])
df['On_Call_Offer']=lab.fit_transform(df['On_Call_Offer'])
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
df['On_Call_Offer'].unique()
```

```python
ann=Sequential()
ann.add(Dense(7,activation='relu'))
ann.add(Dense(10,activation='relu'))
ann.add(Dense(10,activation='relu'))
ann.add(Dense(5,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
ann.fit(x_train,y_train,epochs=100,batch_size=200)
```

```python
ann.evaluate(x_test,y_test)
```
