```python
import cv2
import mediapipe as mp
```

```python
mphands=mp.solutions.hands
mpdraw=mp.solutions.drawing_utils
data=mphands.Hands(max_num_hands=1)
```

```python

Video=cv2.VideoCapture(0)
while True:
    suc,img=Video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    # print(result.multi_hand_landmarks)
    tipid=[4,8,12,16,20]
    lmlst=[]
  
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            for id,lm in enumerate(x.landmark):
                # print(id,lm)
                cx=lm.x
                cy=lm.y
                lmlst.append([id,cx,cy])
                if len(lmlst)!=0 and len(lmlst)==21:
                    fingercount=[]
                    if lmlst[12][1]<lmlst[20][1]:
                        if lmlst[4][1]>lmlst[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    else:
                        if lmlst[4][1]<lmlst[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)

                    #for fingers exceptthumb
                    for i in range(1,5):
                        if lmlst[tipid[i]][2]>lmlst[tipid[i]-2][2]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    # print(fingercount)
                    count=sum(fingercount)
                    cv2.putText(img,"Count:"+str(count),(35,150),cv2.FONT_HERSHEY_COMPLEX,3,(0,0,255),3)



            mpdraw.draw_landmarks(img,x,mphands.HAND_CONNECTIONS,mpdraw.DrawingSpec(color=(255,0,0),thickness=5))
    
    
    
    
    cv2.imshow('Image',img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
Video.release()
cv2.destroyAllWindows()
```
