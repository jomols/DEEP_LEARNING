```python
import cv2
import mediapipe as mp
```

```python
mphands=mp.solutions.hands
mpdraw=mp.solutions.drawing_utils
data=mphands.Hands(max_num_hands=1)
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    tipid=[4,8,12,16,20]
    lmlst=[]
    gesture="" 

    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            lmlst=[] 
            for id,lm in enumerate(x.landmark):
                cx=lm.x
                cy=lm.y
                lmlst.append([id,cx,cy])

                if len(lmlst)==21:
                    # if lmlst[12][1]<lmlst[20][1]:
                    if lmlst[12][2] > lmlst[9][2] and lmlst[16][2] > lmlst[13][2] and lmlst[8][2] < lmlst[6][2] and lmlst[20][2] < lmlst[18][2] and lmlst[4][1]>lmlst[3][1]:
                        gesture="i love you"
                    
                    elif lmlst[12][2] > lmlst[9][2] and lmlst[16][2] > lmlst[13][2] and lmlst[8][2] < lmlst[6][2] and lmlst[20][2] < lmlst[18][2]:
                        gesture="rock on" 

                    elif  lmlst[4][1]<lmlst[2][1] and lmlst[16][2] > lmlst[13][2] and lmlst[20][2] > lmlst[17][2] and lmlst[8][1]==lmlst[12][1] and lmlst[8][2]==lmlst[12][2]:
                        gesture = "good luck"


                    elif lmlst[8][2] < lmlst[6][2] and lmlst[12][2] < lmlst[10][2] and lmlst[16][2] > lmlst[13][2] and lmlst[20][2] > lmlst[17][2]:
                        gesture = "Victory"

                    elif lmlst[4][2] < lmlst[2][2] and lmlst[20][1] > lmlst[18][1] and lmlst[16][1]<lmlst[14][1] and lmlst[8][1] < lmlst[6][1] and lmlst[12][1] < lmlst[10][1]:
                        gesture = "Call Me"

                    fingers_closed = lmlst[8][1] < lmlst[6][1] and lmlst[12][1] < lmlst[10][1] and lmlst[16][1] < lmlst[14][1] and lmlst[20][1] < lmlst[18][1]

                    if lmlst[4][2] < lmlst[2][2] and fingers_closed:
                        gesture = "Thumbs Up"

                    elif lmlst[4][2] > lmlst[2][2] and fingers_closed:
                        gesture = "Thumbs Down"

                    elif lmlst[4][1]>lmlst[2][1] and lmlst[8][2] < lmlst[6][2] and lmlst[12][2] > lmlst[10][2] and lmlst[16][2] > lmlst[14][2] and lmlst[20][2] > lmlst[18][2]:
                        gesture = "Pointing Up"

                    elif lmlst[8][2]<lmlst[6][2] and lmlst[12][2]<lmlst[10][2] and lmlst[16][2]<lmlst[14][2] and lmlst[20][2]<lmlst[18][2] and lmlst[4][2]<lmlst[2][2]:
                        gesture="stop"
                
                    elif lmlst[12][2] < lmlst[10][2] and lmlst[16][2] < lmlst[14][2] and lmlst[20][2] < lmlst[18][2]:
                        dist = ((lmlst[4][1] - lmlst[8][1])**2 + (lmlst[4][2] - lmlst[8][2])**2)**0.5
                        if dist < 40: # If tips are close together
                            gesture = "Super"

                    

            mpdraw.draw_landmarks(img,x,mphands.HAND_CONNECTIONS)
    
    
    if gesture != "":
        cv2.putText(img,gesture,(35,150),cv2.FONT_HERSHEY_COMPLEX,3,(0,0,255),3)
        # print(gesture)

    cv2.imshow("image",img)
    if cv2.waitKey(1)&0XFF==ord("q"):
        break

video.release()
cv2.destroyAllWindows()
```
