```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("WhatsApp Image 2026-02-06 at 11.25.17 AM.jpeg")
```

```python
convert=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(convert)
```

```python
eye_data=cv2.CascadeClassifier("haarcascade_eye.xml")
eye_data
```

```python
eyes=eye_data.detectMultiScale(convert,minNeighbors=5)
eyes
```

```python

for (x,y,w,h) in eyes:
    cv2.rectangle(convert,(x,y),(x+w,y+h),(255,0,0),3)

```
