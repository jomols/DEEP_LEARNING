```python
import matplotlib.pyplot as plt
import cv2
img=cv2.imread("group2.jpeg")
```

```python
convert=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(convert)
```

```python
face_data=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face_data
```

```python
faces=face_data.detectMultiScale(convert,minNeighbors=3)
faces
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(convert,(x,y),(x+w,y+h),(255,0,0),3)
```
