```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("virat_abd.jpeg")
```

```python
convert=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(convert)
```

```python
#face detection
face_data=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face_data
```

```python
#detect multiscale
faces=face_data.detectMultiScale(convert)
faces
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(convert,(x,y),(x+w,y+h),(255,0,0),3)
```
