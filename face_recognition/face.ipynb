```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
clssnames=[]
for i in mylist:
    imgpath=os.path.join(path,i)
    img1=cv2.imread(imgpath)
    imgs.append(img1)
    clssnames.append(i.split(".")[0]) #or we can use i[-4] for all are in jpg format or we can use i[-5] for all are in jpeg format
```

```python
def faceencodings(images):
    encodelist=[]
    for i in imgs:
        img1=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img1)
        # print(face_in_frame)
        face_encode=fr.face_encodings(img1,face_in_frame)[0]
        encodelist.append(face_encode)
    return encodelist
```

```python
encode_list_known_faces=faceencodings(imgs)
```

```python
import pyttsx3
txt_sp=pyttsx3.init()
voices = txt_sp.getProperty('voices')
#predict live face from known faces
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_loc1=fr.face_locations(img1)
    # print(face_loc1)
    faceencode1=fr.face_encodings(img1,face_loc1)
    for enc_loc,faceloc in zip(faceencode1,face_loc1):
        matches=fr.compare_faces(encode_list_known_faces,enc_loc)
        # print(matches)
        face_dis=fr.face_distance(encode_list_known_faces,enc_loc)
        # print(face_dis)
        matchindex=np.argmin(face_dis)
        # print(matchindex)
        if matches[matchindex]==True:
            name=clssnames[matchindex]
            #print(name)

            y1,x2,y2,x1=faceloc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),2)
            cv2.putText(img,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),3)
        else:
            name="unknown"
            y1,x2,y2,x1=faceloc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),2)
            cv2.putText(img,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),3)

        txt_sp.setProperty('voice', voices[1].id)
        txt_sp.setProperty('volume', 1.0)
        txt_sp.setProperty('rate', 150)
        txt_sp.say(name)
        txt_sp.runAndWait()


    cv2.imshow("frames",img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
