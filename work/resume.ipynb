```python
keyword=["Scikit-learn","Pandas","NumPy","Matplotlib","PyTorch","TensorFlow","Machine Learning","Deep Learning","Artificial Inteligence","Data Science","Python","feature engineering","data cleaning"]
```

```python
count=0
for i in range(n):
    if any(k in dic['text'][i] for k in keyword):
        count+=1
        print(dic['text'][i])
        x=dic['left'][i]
        y=dic['top'][i]
        w=dic['width'][i]
        h=dic['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imshow('Frame',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
