```python
import cv2
```

```python
img=cv2.imread("WhatsApp Image 2026-02-06 1.jpeg")
img
```

```python
import matplotlib.pyplot as plt
convert=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(convert)
```

```python
cv2.rectangle(convert,pt1=(280,100),pt2=(680,500),color=(255,0,0),thickness=3)
plt.imshow(convert)
```
